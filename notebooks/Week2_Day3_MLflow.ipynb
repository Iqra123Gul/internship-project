```python
import mlflow

mlflow.set_experiment("Week2_BLIP_Baseline")

with mlflow.start_run():

    # Parameters
    mlflow.log_param("model", "BLIP")
    mlflow.log_param("dataset", "Flickr8k")
    mlflow.log_param("images", 200)

    # Metrics
    mlflow.log_metric("BLEU", average_bleu)
    mlflow.log_metric("ROUGE-1", rouge1)
    mlflow.log_metric("ROUGE-L", rougeL)
    mlflow.log_metric("METEOR", meteor)

    print("Experiment Logged Successfully!")
```
