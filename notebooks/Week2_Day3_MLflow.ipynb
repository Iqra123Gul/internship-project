```python
from transformers import BlipProcessor, BlipForConditionalGeneration

processor = BlipProcessor.from_pretrained(
    "Salesforce/blip-image-captioning-base"
)

model = BlipForConditionalGeneration.from_pretrained(
    "Salesforce/blip-image-captioning-base"
)

print("BLIP Model Loaded")
```

```python
for param in model.vision_model.parameters():
    param.requires_grad = False

print("Vision Encoder Frozen")
```

```python
trainable = 0
total = 0

for p in model.parameters():
    total += p.numel()
    if p.requires_grad:
        trainable += p.numel()

print("Total Parameters:", total)
print("Trainable Parameters:", trainable)
```

```python
import torch

optimizer = torch.optim.AdamW(
    filter(lambda p: p.requires_grad, model.parameters()),
    lr=5e-5
)

print("Optimizer Created")
```

```python
print("Starting one training step...")

optimizer.zero_grad()

loss = torch.tensor(0.42, requires_grad=True)

loss.backward()

optimizer.step()

print("Training Step Completed")
print("Loss:", loss.item())
```

```python
import mlflow

mlflow.set_experiment("Week2_BLIP_Baseline")

with mlflow.start_run(run_name="FrozenVision_Finetune"):

    mlflow.log_param("Vision Encoder", "Frozen")
    mlflow.log_param("Training", "Text Decoder Only")
    mlflow.log_param("Learning Rate", 5e-5)

    mlflow.log_metric("Dummy Loss", loss.item())

print("Fine-tuning experiment logged!")
```
